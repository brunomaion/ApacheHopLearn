```python
import pandas as pd

# Assuming you want to read a CSV file into a pandas DataFrame
df = pd.read_csv('Bakery Sales.csv')
display(df)
df.info()
```

```python

dfProdutos = df


dfProdutos = dfProdutos.drop(columns=['total', 'place'])

dfProdutos_melt = dfProdutos.melt(id_vars=['datetime', 'day of week'], var_name='product', value_name='quantity')

dfProdutos_melt = dfProdutos_melt.dropna(subset=['quantity']).sort_values(by=['datetime'])
display(dfProdutos_melt)

dfProdutos_melt.to_csv('vendas.csv', index=False)
```
